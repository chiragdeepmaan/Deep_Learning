# Feature Engineering
Process of creating, selecting, and transforming features to improve model performance.

Converts raw data into meaningful inputs for machine learning.

Helps models learn patterns more effectively.

Often requires domain knowledge and data analysis.

**Common Feature Engineering Techniques**


*   Feature Creation – Create new features from existing ones (e.g., Age from Date of Birth).
*   Feature Transformation – Scale, normalize, or log-transform data.


*   Feature Encoding – Convert categorical data into numerical form (One-Hot Encoding, Label Encoding).
*   Feature Selection – Keep only the most important features and remove irrelevant ones.
*   Handling Missing Values – Fill or remove missing data.
*   Feature Extraction – Reduce dimensions while retaining important information (e.g., PCA).





**Why is Feature Engineering Important?**

Improves model accuracy.

Reduces overfitting.

Speeds up model training.

Makes models more robust and interpretable.

```python
# standardization
data['Age'].plot(kind='hist')
plt.show()
sns.kdeplot(data['Age'])
#
```

```python
data['Salary'].plot(kind='hist')
plt.show()
sns.kdeplot(data['Salary'])
```
